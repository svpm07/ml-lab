```python
x=df[["GRE Score","TOEFL Score","CGPA"]]
y=df["Chance of Admit"]
```

```python
x
```

```python
y
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=3)
```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
xtest.shape
```

```python
ytest.shape
```

```python
y_train=np.array([1 if y > 0.82 else 0 for y in ytrain])
y_test=np.array([1 if y > 0.82 else 0 for y in ytest])
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_model=DecisionTreeClassifier(criterion="gini",max_depth=3,random_state=3)
```

```python
tree_model.fit(xtrain,y_train)
```

```python
y_pred=tree_model.predict(xtest)
```

```python
y_pred
```

```python
from sklearn import metrics
```

```python
print(metrics.confusion_matrix(y_test,y_pred))
```

```python
confusion_matrix=metrics.confusion_matrix(y_test,y_pred)
cm_display=metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix)
cm_display.plot()
```

```python
print("Accuracy =",metrics.accuracy_score(y_test,y_pred))
```

```python
print("Precision =",metrics.precision_score(y_test,y_pred))
```

```python
print("Recall =",metrics.recall_score(y_test,y_pred))
```

```python
print("F1 Score =",metrics.f1_score(y_test,y_pred))
```

```python
from sklearn import tree
tree.plot_tree(tree_model)
```
