```python
product=[]
for i in range(0,7501):
    product.append([str(df.values[i,j])for j in range(20)])
    
```

```python
print(product[2])
```

```python
model=apriori(product,min_support=.003,min_confidense=.3,min_lift=3,min_length=2,max_length=2)
```

```python
results=list(model)
for i in results:
    print('\n')
    print(i)
    print("------------------------------------")
```
