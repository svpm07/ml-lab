```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn import linear_model
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x=df[["YEAR"]]
```

```python
y=df[["JAN"]]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.25,random_state=2)
```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
xtest.shape
```

```python
ytest.shape
```

```python
reg=linear_model.LinearRegression()
model=reg.fit(xtrain,ytrain)
```

```python
print("Predicted temp of JAN 2025 :",reg.predict([[2025]]))
```

```python
y_pred=model.predict(xtest)
y_pred
```

```python
print("Slope of the model",model.coef_)
```

```python
print("Intercept of the model",model.intercept_)
```

```python
import numpy as np
from sklearn import metrics
mae=metrics.mean_absolute_error(ytest,y_pred)
mse=metrics.mean_squared_error(ytest,y_pred)
r2=metrics.r2_score(ytest,y_pred)
rmse=np.sqrt(mse)
```

```python
print("Mean Absolute Error =",mae)
print("Mean Squared Error =",mse)
print("R2 Score =",r2)
print("Root Mean Square Error =",rmse)
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.scatter(xtest,ytest,label="Actual Data",color='red')
plt.plot(xtest,y_pred,label="Regression Line",color="blue",linewidth=2)
plt.title("Year vs Jan Temp in 'C")
plt.xlabel("Year")
plt.ylabel("Temp in Jan ")
plt.legend()
plt.show()
```
